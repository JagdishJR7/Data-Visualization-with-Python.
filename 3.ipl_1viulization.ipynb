```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
sns.set_style("darkgrid")
ls=match_df['venue'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(15,4))
temp =plt.bar(ls.index, ls.values, alpha=0.4)
plt.title('MOST PLAYED VENUE')
plt.ylabel('COUNT', fontsize=14)
plt.xlabel('NAME OF THE STADIUMS', fontsize=5)
temp.set_xticklabels(rotation=0,labels=ls.index,fontsize=1)
plt.show()
```
