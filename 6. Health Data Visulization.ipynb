# Percentaze wise Diabetese Patient:

```python
plt.pie(df["Diabetes"].value_counts(),labels = df["Diabetes"].unique(),autopct = "%.2f%%",shadow=True,explode=(0,0.3),radius=1.5);
plt.title("Diabetes Patient Or Not")
plt.legend()
plt.show()
```

```python
plt.pie(df["Hypertension"].value_counts(),labels = df["Hypertension"].unique(),autopct = "%.2f%%",shadow=True,explode=(0,0.3),radius=1.5);
plt.title("Hypertension patient Or Not")
plt.legend()
plt.show()
```

```python
plt.pie(df["Stroke"].value_counts(),labels = df["Stroke"].unique(),autopct = "%.2f%%",shadow=True,explode=(0,0.3),radius=1.5);
plt.title("Stroke patient Or Not")
plt.legend()
plt.show()
```
