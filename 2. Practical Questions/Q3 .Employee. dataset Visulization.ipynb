## Q3 . i) Draw scatter plot for age and salary:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.relplot(x = 'age', y = 'salary',kind= "scatter",data = data)
```

## ii) Plot histogram for features age and salary:


```python
plt.hist(data['age'])
plt.title("Hisogram Chart")
plt.xlabel('age')
plt.show()
```

```python
plt.hist(data['salary'])
plt.title("Hisogram Chart")
plt.xlabel('salary')
plt.show()
```
